## Import the data

```python
img_path = r'C:\Users\asiaynrf\Desktop\fun\resource\datasets\flower\flower_photos'
```

```python
flower_class = []
for dirs in os.listdir (img_path):
    if os.path.isdir (os.path.join (img_path, dirs)):
        flower_class.append (str(dirs))
flower_class_dir = {}
for i in range (len(flower_class)):
    flower_class_dir [flower_class[i]] = i
```

## Attach the file_path with the class

```python
X, y = [], []
check = []
for curr_class in flower_class:
    curr_dir = os.path.join (img_path, curr_class)
    counter = 0
    for filepath in os.listdir (curr_dir):
        counter += 1
        if filepath.endswith (".jpg"):
            X.append (os.path.join (curr_dir, filepath))
            y.append (flower_class_dir[curr_class])
    print ('curr checking %s and there is %d images' % (curr_class, counter))
```

## Split into train_validation_test

```python
N = len(X) 
validation_ratio = 0.05
test_ratio = 0.2
```

```python
z = list(zip (X, y))
```

```python
import random
random.shuffle (z)
X, y = zip(*z)
test_threshold  = int (np.ceil (test_ratio * N))
validation_threshold = test_threshold +  int (np.ceil (validation_ratio * N))
print (test_threshold, validation_threshold)
X_test, X_validation, X_train = X[0: test_threshold],X [test_threshold: validation_threshold], X[validation_threshold:]
y_test, y_validation, y_train = y[0: test_threshold],y [test_threshold: validation_threshold], y[validation_threshold:]
print (len(X_test), len(X_validation), len(X_train), N)
print (Counter (y_test), Counter (y_validation))
```

## Get the network from tensorflow library and test with the known picture

```python
from tensorflow.contrib.slim.nets import inception
import tensorflow.contrib.slim as slim
```

```python
height, width, channel = 299, 299, 3
tf.reset_default_graph()
```

```python
X_tf = tf.placeholder (tf.float32, shape = [None, height, width, channel], name = 'X')
y_tf = tf.placeholder (tf.int32, shape = [None], name = 'y')
```

```python
with slim.arg_scope (inception.inception_v3_arg_scope ()):
    logits, end_point = inception.inception_v3 (X_tf, num_classes = 1001, is_training= False)
predictions = end_point["Predictions"]
saver = tf.train.Saver()
```

```python
check_point = r"C:\Users\asiaynrf\Desktop\fun\resource\datasets\inception\inception_v3.ckpt"
```

```python
import matplotlib.image as mping
X_example = mping.imread (r'C:\Users\asiaynrf\Desktop\fun\resource\image\download.png')
X_example = X_example *2 - 1
X_example = X_example.reshape (-1, 299, 299, 3)
```

```python
with tf.Session () as sess:
    saver.restore (sess, check_point)
    predictions_val = sess.run (predictions, feed_dict = {X_tf: X_example})
```

```python
most_likely_class = np.argmax (predictions_val)
```

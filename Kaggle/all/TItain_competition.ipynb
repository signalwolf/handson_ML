# Clear the data

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
# drop the Name and Ticket info, not needed
train_data = train_data.drop("Name", axis = 1)
train_data = train_data.drop("Ticket", axis = 1)
train_data.head()
```

```python
# from below information, there is few thing needed to change before train:
# 1.In Age, there is 891 - 714 Null data. Change it to average
# 2.Sex needed to convert to 0/1
# 3.Cabin/Embarked needed to convert to int
# 4.Cabin's Non shall keep there and convert to int
# 5.Embarked have two Non data, drop it.
```

```python
# fill NA in Age to it's median
median = train_data["Age"].median()
train_data["Age"].fillna(median, inplace=True) # option 3
train_data.info()
```

```python
# Drop NA in Embarked:
train_data = train_data.dropna(axis=0,subset=["Embarked"])
train_data.info()
```

```python
train_data.head()
```

```python
# Encode the sex and Embard
#ordinal_encoder = factorize()
train_data["Sex"], category = train_data["Sex"].factorize()
train_data["Embarked"], category = train_data["Embarked"].factorize()
train_data.head()
```

```python
train_data["Cabin"].fillna(0, inplace=True)
```

```python
train_data["Cabin"], cat = train_data["Cabin"].factorize()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(train_data, test_size=0.1)
```

```python
train_label, test_label = train_set["Survived"], test_set["Survived"]
train_set, test_set = train_set.drop("Survived", axis = 1), test_set.drop("Survived", axis = 1)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(train_set, train_label)

```

```python
from sklearn.metrics import mean_squared_error
test_predict = lin_reg.predict(test_set)
lin_mse = mean_squared_error(test_label, test_predict)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(train_set, train_label)
```

```python
from sklearn.metrics import mean_squared_error
test_predict = tree_reg.predict(test_set)
lin_mse = mean_squared_error(test_label, test_predict)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.svm import SVR

svm_reg = SVR(kernel="linear")
svm_reg.fit(train_set, train_label)
test_predict = svm_reg.predict(test_set)
svm_mse = mean_squared_error(test_label, test_predict)
svm_rmse = np.sqrt(svm_mse)
svm_rmse
```
